```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure


#Read in the data
df = pd.read_csv(r'C:\Users\Max Causso\Downloads\archive3\movies.csv')

```

```python
#Let's look at the data
df.head()
```

```python
#Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#Data types for out columns

df.dtypes
```

```python
#change data type type of a columns from float to int

def convertirbudget(valor):
    if pd.isna(valor):
        return 0
    return int(valor)

df['budget'] = df['budget'].apply(convertirbudget)

def convertirgross(valor):
    if pd.isna(valor):
        return 0
    return int(valor)

df['gross'] = df['gross'].apply(convertirgross)
```

```python
df
```

```python
#create correct YEAR column
df['yearcorrect'] = df['released'].astype(str).str[:4]
df
```

```python
#Ordenar segun el Gross de mayor a menor
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#Numerized, volver en numeros los que son de algun tipo;
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
